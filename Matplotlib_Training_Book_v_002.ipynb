3.5 Plot all (x, np.sin(x-0)) (x, np.sin(x-1)) ... (x, np.sin(x-5)) using following colors for the line:
    
    "blue" specify color by name
    "g" short color code (rgbcmyk)
    "0.75" Grayscale between 0 and 1
    "#FFDD44" Hex code (RRGGBB from 00 to FF)
    (1.0, 0.2 , 0.3) RGB tuple, values 0 and 1
    "chartreuse" all HTML color names supported

```python
plt.plot(x, np.sin(x-0), color= "blue")
plt.plot(x, np.sin(x-1), color= "g")
plt.plot(x, np.sin(x-2), color= "0.75")
plt.plot(x, np.sin(x-3), color= "#FFDD44")
plt.plot(x, np.sin(x-4), color= (1.0,0.2,0.3))
plt.plot(x, np.sin(x-5), color= "chartreuse")
```

3.12 Plot the sin curve as a solid green line, plot the cos curve as a dotted blue line. Add a legend with "sin(x)" and "cos(x)". Set the x-axis equal to the y-axis

```python
plt.plot(x, np.sin(x), linestyle= "solid", color= "green", label= "sin(x)")
plt.plot(x, np.cos(x), linestyle= "dotted", color= "blue", label= "cos(x)")
plt.axis("equal")

plt.legend()
```
